```python
#linear regression only for linear data, not exponential


from statistics import mean
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import random

style.use('fivethirtyeight')
```

```python
xs = np.array([1,2,3,4,5,6], dtype=np.float64)
ys = np.array([4,7,3,2,6,8], dtype=np.float64)

plt.scatter(xs,ys)
plt.show()
```

```python
# best fit slop and intercept
def bfsai(xs,ys):
    m = ( mean(xs)*mean(ys) - mean(xs*ys) ) / ( mean(xs)**2 - mean(xs**2) )
    b = ( mean (ys) - m*(mean(xs)))
    return m, b
```

```python
m,b= bfsai(xs,ys)
print(m,b)
```

```python
 # equation of a line is y = mx + b
    
regression_line = [(m*x)+b for x in xs]
```

```python
#The best fit line

plt.scatter(xs,ys)
plt.plot(xs, regression_line)
plt.show()
```

```python
predict_x = 8
predict_y = m*predict_x + b
```

```python
plt.scatter(xs,ys)
plt.scatter(predict_x,predict_y)
plt.plot(xs, regression_line)
plt.show()
```

```python
# how good is the best fit line?
# using squared error
# error is the distance from the data point 
# coefficient of determination 
```

### Coefficient of Detemination

```python
def squared_error(ys_org, ys_line):
    return sum((ys_line-ys_org)**2)

def coefficient_of_determination(ys_org, ys_line):
    y_mean_line = [mean(ys_org) for y in ys_org]
    squared_error_regr = squared_error(ys_org, ys_line)
    squared_error_y_mean_line = squared_error(ys_org, y_mean_line)
    return 1 - (squared_error_regr / squared_error_y_mean_line)

```

```python
r_squared = coefficient_of_determination(ys, regression_line)
```

```python
print(r_squared)
```

## REGRESSION MODEL

```python
def linear_regression(x1s,y1s):
    
    # best fit slop & the intercepy
    m1 = ( mean(x1s)*mean(y1s) - mean(x1s*y1s) ) / ( mean(x1s)**2 - mean(x1s**2) )
    b1 = ( mean (y1s) - m1*(mean(x1s)))
    
    # plotting the regression line
    regression_line_1 = [(m1*x)+b1 for x in x1s]
    
    # coeffecient of determination
    r_squared = coefficient_of_determination(y1s, regression_line_1)
    return r_squared
```

```python
def create_dataset(hm, variance, step=2, correlation=False):
    val = 1
    ys = []
    for i in range(hm):
        y = val + random.randrange(-variance, variance)
        ys.append(y)
        if correlation and correlation == 'pos':
            val+=step
        if correlation and correlation == 'neg':
            val-=step 
        xs = [i for i in range (len(ys))]
    return np.array(xs, dtype=np.float64), np.array(ys,dtype=np.float64)
```
